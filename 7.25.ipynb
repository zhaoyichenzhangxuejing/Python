```python
#按照升序排列
b = [4,3,2,1]
length = len(b)
for i in range(length):
    for j in range(length-i-1):  #一次锁定一个最大值
        if b[j] > b[j+1]:
            b[j],b[j+1] = b[j+1],b[j]
b
```
