## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

```python
i = 0
while i < 10:
    print('hahaha') # 查看打了多少次，做题时也可用此进行判断！
    i += 1 # 自增1 ！
```

# 验证码
- 随机产生四个字母的验证码，如果正确，输出验证码正确。如果错误，产生新的验证码，用户重新输入。
- 验证码只能输入三次，如果三次都错，返回“别爬了，我们小网站没什么好爬的”
- 密码登录，如果三次错误，账号被锁定


```python
import random
```

```python
x = random.randint(1000,9999)
print(x)
y = eval(input('输入四个数字'))
i = 1
while i < 3:
    if x == y:
        print('right')
        break
    else:
        x = random.randint(1000,9999)
        print(x)
        y = eval(input('输入四个数字'))
        i += 1
```

```python
x = random.randrange(aaaa,bbbb)
print(x)
y = eval(input('输入四个字母'))
i = 1
while i < 3:
    if x == y:
        print('right')
        break
    else:
        x = random.randint(aaaa,bbbb)
        print(x)
        y = eval(input('输入四个字母'))
        i += 1
```

## for 循环
- Python的for 循环通过一个序列中的每个值来进行迭代
- range(a,b,k)， a，b，k 必须为整数
- a: start
- b: end
- k: step
- 注意for 是循环一切可迭代对象，而不是只能使用range

```python
for i in range(1,10,2):#1-10内部长为2
    print(i)
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

## EP：
- 使用多层循环完成9X9乘法表
- 显示50以内所有的素数

```python
for i in range(1,10):
    for j in range(1,i+1):
        print(j,'x',i,'=',i*j,end = '   ')
    print()
```

```python
i = 2
while (i < 50):
    j = 2
    while (j <= (i/j)):
        if not (i%j):
            break
        j = j + 1
        if (j > i/j):
            print(i,' ',end = ' ')
        i = i + 1
```

## 关键字 break 和 continue
- break 跳出循环，终止循环
- continue 跳出此次循环，继续执行

```python
for i in range(10):
    if i == 5:
        break
print(i)
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(10):
    if i == 5:
        pass
    else:
        print(i)
```

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            print(i,j,k)
```

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            if k == 5:
                continue
            print(i,j,k)
```

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            if k == 5:
                break
            print(i,j,k)
```
